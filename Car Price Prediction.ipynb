```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data['Age'] = 2020 - data['Year']
```

```python
data = data.drop(['Car_Name','Year'],axis = 'columns')
```

```python
data.head()
```

```python
data = pd.get_dummies(data,drop_first = True)
data.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize = (20,20))
g = sns.heatmap(data.corr(),annot = True,cmap = 'coolwarm')
```

```python
data.corr()['Selling_Price']
```

```python
X = data.iloc[:,1:]
Y = data.iloc[:,0]
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,Y)
```

```python
x = model.feature_importances_
feat = pd.Series(x,index = X.columns)
feat
```

```python
feat.nlargest(5).plot(kind = 'barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {'n_estimators':(range(100,1200,100)), 
              #'criterion':('mse','mae'),
              'max_features':('auto','sqrt'),
              'max_depth':(range(5,30,5)),
                'min_samples_split': [2,5,10,15,100],
                'min_samples_leaf':[1,2,5,10]}
```

```python
gs = RandomizedSearchCV(rf_regressor,parameters,scoring = 'neg_mean_absolute_error',cv = 3,random_state = 42)
gs.fit(x_train,y_train)
```

```python
gs.best_score_
```

```python
gs.best_estimator_
```

```python
y_pred = gs.predict(x_test)
y_pred
```

```python
sns.distplot(y_test-y_pred)
plt.show()
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression,Lasso
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred2 = regressor.predict(x_test)
```

```python
print(mean_absolute_error(y_test,y_pred2))
```
